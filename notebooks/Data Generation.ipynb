```python
import numpy as np

word_subset = ["ABOUT","BACK","THE","SO","YES"]

data_path = os.path.join(proj_data_dir, "data_gen_1")
word_dirs = os.listdir(path=data_path)

data = []
word_labels = []

files = []
for word_dir in word_subset:
  for file in os.listdir(os.path.join(data_path,word_dir)):
    filename = file.split(".")[0].replace("_","/")
    files.append(os.path.join(f"{proj_data_dir}/LRS2/pretrain",filename +".txt"))
    files.append(os.path.join(f"{proj_data_dir}/LRS2/pretrain",filename+".mp4"))
files = np.unique(files)
print(files)
```
